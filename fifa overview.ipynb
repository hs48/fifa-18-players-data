```python
#rows with balance = NULL
msno.matrix(fifa[fifa["Balance"].isnull()])
```

```python
#drop rows where balance is null
fifa =fifa[~fifa["Balance"].isna()]
```

```python
#delete columns with photo URL
import numpy as np

objects = [x for x in fifa.columns if fifa[x].dtype==object]

ls = []

for i in objects:
    try:
        if fifa.iloc[0][i].startswith('https:'):
            ls.append(i)
    except AttributeError: 
        break
ls
```

```python
fifa1 = fifa
```

```python
#drop URLs
fifa1.drop(ls , axis=1, inplace=True)
```

```python
#view the dataset
pd.set_option('display.max_columns', None)
fifa1.head()
```

```python
#export final csv file
fifa1.to_csv("fifa2.csv")
```
